```python
# Clean data
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score, recall_score, f1_score
# Load the data from CSV
df = pd.read_csv('/content/hospital1.csv')
total_records = df.shape[0]
percent_30 = int(0.30 * total_records)

# Separate training and test data
data_inputs = np.array(df.drop('Diabetic', 1).iloc[percent_30:, :])
data_outputs = np.array(df['Diabetic'].iloc[percent_30:])


# Load test data
test_inputs = np.array(df.drop('Diabetic', 1).iloc[:percent_30, :])
test_outputs = np.array(df['Diabetic'].iloc[:percent_30])


# Scale the data
sc = StandardScaler()
data_inputs = sc.fit_transform(data_inputs)
test_inputs = sc.transform(test_inputs)

# Train RandomForestClassifier
clf = DecisionTreeClassifier(criterion='entropy', max_depth=5,random_state=42)
clf.fit(data_inputs, data_outputs)

# Predict on test data
y_pred = clf.predict(test_inputs)

# Evaluate model performance
rf_acc = '{:.2f}'.format(accuracy_score(test_outputs, y_pred))
print('Random Forest Classifier output on test set: ', rf_acc)
rf_pre = '{:.2f}'.format(precision_score(test_outputs, y_pred))
print('Random Forest Classifier output on test set: ', rf_pre)
rf_rec = '{:.2f}'.format(recall_score(test_outputs, y_pred))
print('Random Forest Classifier output on test set: ', rf_rec)
rf_f1 = '{:.2f}'.format(f1_score(test_outputs, y_pred))
print('Random Forest Classifier output on test set: ', rf_f1)
```
